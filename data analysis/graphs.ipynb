## Chart and graphs
Types of graph
- üìä Bar chart : used to compare the values of different categories of data
- üìà Line chart : used to show trends over time
- üìâ Area chart : used to show trends over time, but can also be used to show the difference between two values
- üç∞ Pie chart : used to show the proportion of each category of data
- üìä Histogram : used to show the distribution of a numerical variable
- üìà Scatter plot : used to show the relationship between two numerical variables

````python
rainfall = np.array([ 399, 555,  464,  587,  559])
years = np.array([2019, 2020, 2021, 2022, 2023])
````

````python
# line plot
plt.plot(rainfall) 
````

````python
plt.plot(years, rainfall)
````

````python
# configuring graph
plt.plot(years, rainfall)
plt.title("Rainfall per year", fontdict={'fontsize':15,'color':'green'})
plt.xlabel("Years")
plt.ylabel("Rainfall (cm)")
plt.xticks(years)
plt.hlines(y=500, xmin=2019, xmax=2023, color='red', linestyles='--')
plt.show()
````

````python
plt.rcParams['font.family'] = 'Copperplate Gothic Bold'
plt.figure(figsize=(10, 4), facecolor='yellow')
x = np.array(['Rohan', 'Rahul', 'Raj', 'Ramesh', 'Rakesh'])
y = np.array([60, 80, 90, 20, 50])
bars = plt.bar(x, y, 
        color=['orange','green','red','blue','pink'],
        alpha=.5, edgecolor='black',
        linewidth=1, hatch='/', width=.5)
bars[2].set_hatch('o')
bars[2].set_color('red')
plt.title("Data")
plt.ylim(0, 120)
for bar in bars:
    plt.text(bar.get_x()+bar.get_width()/2, 
             bar.get_height()+10, 
             f"${bar.get_height()}k",
             fontsize=10, color='red',
             ha='center')
plt.show()
````

````python
plt.figure(figsize=(5, 5), facecolor='white')
plt.pie(
    y, 
    labels = x,
    explode = [0, 0, .05, 0, 0],
    autopct='%.1f%%',
    pctdistance=.75,
    counterclock=True,
    startangle=90,
    wedgeprops={'width':.5, 'edgecolor':'white'}, # donut chart
)
plt.show()
````
